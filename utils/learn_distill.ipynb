```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lg = np.random.uniform(low=1, high=20,size=10)
```

```python
def plt_sfmx(lg, T=1):
    lg_ = lg/T
    cum_lg = np.sum(np.exp(lg_))
    cum_lg
    
    plt.plot(range(1,len(lg)+1), [np.exp(_)/cum_lg for _ in lg_])
    plt.xticks([_ for _ in range(1,len(lg)+1)])
    plt.show
```

```python
plt_sfmx(lg,T=1)
```

```python
plt_sfmx(lg,T=2)
```

```python
plt_sfmx(lg,T=5)
```

```python
plt_sfmx(lg,T=10)
```
